# Introduction
Artificial Intelligence is the study of building agents that act rationally. Most of the time, these agents perform some kind of Artificial Intelligence in games goes toward defining the way a computer opponent behaves. Games is a multiagent environments, in which each agent needs to consider the actions of other agents.in this phase of the project, we will consider the alpha-beta algorithm.
<br>
Alpha-beta is a way of finding the optimal minimax solution while avoiding searching subtrees of moves which won't be selected. Where minimax algorithm is a way of finding an optimal move in a two player game.

# Explanation
In the search tree for a two-player game, there are two kinds of nodes:
<br>
1- Nodes representing your moves are called MAX. The goal at a MAX node is to maximize the value of the subtree rooted at that node. To do this, a MAX node chooses the child with the greatest value, and that becomes the value of the MAX node.
<br>
2- Nodes representing your opponent's are called These are also called MIN. The goal at a MIN node is to minimize the value of the subtree rooted at that node. To do this, a MIN node chooses the child with the least value, and that becomes the value of the MIN node.
<br>
Alpha-beta gets its name from two bounds that are passed along during the calculation, which restrict the set of possible solutions based on the portion of the search tree that has already been seen. Specifically, Beta ”β” is the minimum upper bound of possible solutions. Alpha ”α” is the maximum lower bound of possible solutions. Thus, when any new node is being considered as a possible path to the solution, it can only work if:
<br>
α <= N<= β
<br>
where N is the current estimate of the value of the node.

# Evaluation function 
Evaluation function used to evaluate the 􏰀goodness􏰁 of a game position. The zero-sum assumption lets us use one function to describe goodness for both players.<br>
 - f(n)>0 if 'X' win in position n<br>
 - f(n)=0 if position n is tie<br>
 - f(n)<0 if 'O' win in position n

A simple evaluation function for Tic-Tac-Toe without reaching a terminal state if **max is 'X'**: count number of lines where X can win subtract number of lines where O can win.<br>
•  Value of evaluation function at the start of game is zero becuse on an empty game board there are 8 possible winning lines for both X and O as shown in fig1.
<img src="fig1.png" width=300 ><br>
evalX = (number of lines where X can win) – (number of lines where O can win)<br>
•  After X moves in center, score for X is +4 •  After O moves, score for X is +2<br>
•  After X’s next move, score for X is +4<br>
as shown in fig2

**In our problem, we reach a terminal state, if the value of state 1 means the max ('X') win, -1 means the min ('O') win and  0 means tie.**

# Implementation
<h1 dir="rtl">تطبيق الخوارزمية </h1>

```python
from colorama import Fore, Style
#Create class Game
class Game:
    #initialise board of indexes and index dictionary
    indexes = [['1','2','3'],
              ['4','5','6'],
              ['7','8','9']]
    index_dic= {1:(0,0), 2:(0,1), 3:(0,2),
                4:(1,0), 5:(1,1), 6:(1,2),
                7:(2,0), 8:(2,1), 9:(2,2)}
    def __init__(self):
        # Player X always plays first
        self.player_turn = 'X'
        
    #for draw board of indexes
    def draw_indexes(self):
        for i in range(0, 3): 
            for j in range(0, 3):
                if j!= 2:
                    print('{}|'.format(self.indexes[i][j]), end="")
                else :
                    print(self.indexes[i][j],end="")
            if i !=2:
                print('\n-+-+-')
        print('\n')
        
    #for draw board of current state
    def draw_board(self,current_state):
        for i in range(0, 3): 
            for j in range(0, 3):
                if j!= 2:
                    print('{}|'.format(current_state[i][j]), end="")
                else :
                    print(current_state[i][j],end="")
            if i !=2:
                print('\n-+-+-')
        print('\n')
        
    # Determines if the made move is a legal move
    def is_valid(self, px, py,current_state):
        if px < 0 or px > 2 or py < 0 or py > 2:
            return False
        elif current_state[px][py] != ' ':
            return False
        else:
            return True    
        
    # Checks if the game has ended and returns the winner in each case
    def is_end(self,current_state):
        # Vertical win
        for i in range(0, 3):
            if (current_state[0][i] != ' ' and
                current_state[0][i] == current_state[1][i] and
                current_state[1][i] == current_state[2][i]):
                return current_state[0][i]

        # Horizontal win
        for i in range(0, 3):
            if (current_state[i] == ['X', 'X', 'X']):
                return 'X'
            elif (current_state[i] == ['O', 'O', 'O']):
                return 'O'

        # Main diagonal win
        if (current_state[0][0] != ' ' and
            current_state[0][0] == current_state[1][1] and
            current_state[0][0] == current_state[2][2]):
            return current_state[0][0]

        # Second diagonal win
        if (current_state[0][2] != ' ' and
            current_state[0][2] == current_state[1][1] and
            current_state[0][2] == current_state[2][0]):
            return current_state[0][2]

        # Is whole board full?
        for i in range(0, 3):
            for j in range(0, 3):
                # There's an empty field, we continue the game
                if (current_state[i][j] == ' '):
                    return None

        # It's a tie!
        return ' '
    
    # Checks if the game has ended and returns the winner in each case and colored the 3 consecutive marks
    def color_if_end(self,current_state):
        # Vertical win
        for i in range(0, 3):
            if (current_state[0][i] != ' ' and
                current_state[0][i] == current_state[1][i] and
                current_state[1][i] == current_state[2][i]):
                temp=current_state[0][i]

                #color the vertical
                current_state[0][i]=f'{Fore.BLUE}{temp}{Style.RESET_ALL}'
                current_state[2][i]=f'{Fore.BLUE}{temp}{Style.RESET_ALL}'
                current_state[1][i]=f'{Fore.BLUE}{temp}{Style.RESET_ALL}'
                return (temp,current_state)

        # Horizontal win
        for i in range(0, 3):
            if (current_state[i] == ['X', 'X', 'X']):
                #color X horizontal
                current_state[i]=[f'{Fore.BLUE}X{Style.RESET_ALL}',f'{Fore.BLUE}X{Style.RESET_ALL}',f'{Fore.BLUE}X{Style.RESET_ALL}']
                return ('X',current_state)
            elif (current_state[i] == ['O', 'O', 'O']):
                #color O horizontal
                current_state[i]=[f'{Fore.BLUE}O{Style.RESET_ALL}',f'{Fore.BLUE}O{Style.RESET_ALL}',f'{Fore.BLUE}O{Style.RESET_ALL}']
                return ('O',current_state)

        # Main diagonal win
        if (current_state[0][0] != ' ' and
            current_state[0][0] == current_state[1][1] and
            current_state[0][0] == current_state[2][2]):
            temp=current_state[0][0]
            #color the main diagonal
            current_state[0][0]=f'{Fore.BLUE}{temp}{Style.RESET_ALL}'
            current_state[1][1]=f'{Fore.BLUE}{temp}{Style.RESET_ALL}'
            current_state[2][2]=f'{Fore.BLUE}{temp}{Style.RESET_ALL}'
            return (temp,current_state)

        # Second diagonal win
        if (current_state[0][2] != ' ' and
            current_state[0][2] == current_state[1][1] and
            current_state[0][2] == current_state[2][0]):
            temp=current_state[0][2]
            #color the second diagonal
            current_state[0][2]=f'{Fore.BLUE}{temp}{Style.RESET_ALL}'
            current_state[1][1]=f'{Fore.BLUE}{temp}{Style.RESET_ALL}'
            current_state[2][0]=f'{Fore.BLUE}{temp}{Style.RESET_ALL}'
            return (temp,current_state)

        # Is whole board full?
        for i in range(0, 3):
            for j in range(0, 3):
                # There's an empty field, we continue the game
                if (current_state[i][j] == ' '):
                    return (None,current_state)


        # It's a tie!
        return (' ',current_state)

    #max algorithm
    def max_value(self, alpha, beta, current_state):
        #initialize the maxv with minimum value 
        maxv = -2
        px = None
        py = None

        result = self.is_end(current_state)
        #checks if the game has ended
        if result == 'X':
            return (-1, 0, 0)
        elif result == 'O':
            return (1, 0, 0)
        elif result == ' ':
            return (0, 0, 0)
        #if the game has not ended yet search for max value
        for i in range(0, 3):
            for j in range(0, 3):
                if current_state[i][j] == ' ':
                    current_state[i][j] = 'O'
                    (m, min_i, in_j) = self.min_value(alpha, beta, current_state)
                    if m > maxv:
                        maxv = m
                        px = i
                        py = j
                    current_state[i][j] = ' '
                    #cut off
                    if maxv >= beta:
                        return (maxv, px, py)

                    if maxv > alpha:
                        alpha = maxv

        return (maxv, px, py)
    
    #min algorithm
    def min_value(self, alpha, beta,current_state):
        #initialize the minv with maximum value 
        minv = 2
        qx = None
        qy = None

        result = self.is_end(current_state)
        #checks if the game has ended
        if result == 'X':
            return (-1, 0, 0)
        elif result == 'O':
            return (1, 0, 0)
        elif result == ' ':
            return (0, 0, 0)
        #if the game has not ended yet search for min value
        for i in range(0, 3):
            for j in range(0, 3):
                if current_state[i][j] == ' ':
                    current_state[i][j] = 'X'
                    (m, max_i, max_j) = self.max_value(alpha, beta, current_state)
                    if m < minv:
                        minv = m
                        qx = i
                        qy = j
                    current_state[i][j] = ' '
                    #cut off
                    if minv <= alpha:
                        return (minv, qx, qy)

                    if minv < beta:
                        beta = minv

        return (minv, qx, qy)
    #alpha beta algorithm
    def play_alpha_beta(self,current_state):
        print('Position numbers: ')
        self.draw_indexes() 
        print('Game is start!')
        self.player_turn = 'X'
        while True:
            #checks if the game has ended; draw board and show result 
            self.result,current_state = self.color_if_end(current_state)
            self.draw_board(current_state)
            if self.result != None:
                if self.result == 'X':
                    print(f'{Fore.GREEN}The winner is X!🥳')
                elif self.result == 'O':
                    print(f'{Fore.RED}The winner is O!😟')
                elif self.result == ' ':
                    print(f"{Fore.RED}It's a tie!😀")
                return

            if self.player_turn == 'X':
                #ask user play for turn 'X'
                while True:
                    index = int(input("Chose the position number for 'X': "))
                    px,py= self.index_dic[index]
                    
                    if self.is_valid(px, py,current_state):
                        current_state[px][py] = 'X'
                        self.player_turn = 'O'
                        break
                    else:
                        print(f'{Fore.RED}The move is not valid! Try again.{Style.RESET_ALL}')

            else:
                #call max_value for 'O' turn
                print("'O' played :")
                (m, px, py) = self.max_value(-2, 2,current_state)
                current_state[px][py] = 'O'
                self.player_turn = 'X'
```

# Effectiveness of Alpha-Beta Search

Effectiveness Depends On The Order Which successors are examined; more effective if best successors are examined first.

Worst Case:
 - ordered so that no pruning takes place 
 
Best Case:
 - each player’s best move is evaluated first


In practice often get O(b^(d/2)) rather than O(b^d) 



<h1 dir = "rtl">فعالية بحث Alpha-Beta</h1>
<div dir= "rtl">
تعتمد الفعالية على الترتيب الذي يتم فحص الخلفاء منه ؛ أكثر فعالية إذا تم فحص أفضل الخلفاء أولاً.
<br>
الحالة الأسوأ:<br>
  - ترتيب لا يحدث فيه التقليم
 <br>
أفضل حالة:<br>
  - يتم تقييم أفضل حركة لكل لاعب أولاً
  <br>
    عمليًا غالبًا ما نحصل على <span dir= "ltr">O(b^(d/2))</span>
    بدلًا من <span dir= "ltr">O(b^d)</span>
